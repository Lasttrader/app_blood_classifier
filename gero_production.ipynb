## Обработка введенных данных 

```
# StandardScaler + reshape
x_list = []
def get_scaler(xlst, flst):
  flst = array(flst).reshape(-1, 1)
  scaler = StandardScaler()
  x = scaler.fit_transform(flst)
  x = x.reshape(1,-1)
  xlst = xlst.extend(x)
  return xlst
```
